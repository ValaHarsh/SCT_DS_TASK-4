```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Harsh\Achivments\Projects\Skillcraft Tech. internship Project\task 4\US_Accidents_Dec21_updated.csv")

# Display first few rows
df.head()
```

```python
import pandas as pd

# Convert 'Start_Time' to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')

# Extract useful time components
df['Year'] = df['Start_Time'].dt.year
df['Month'] = df['Start_Time'].dt.month
df['Day'] = df['Start_Time'].dt.day
df['Hour'] = df['Start_Time'].dt.hour
df['Weekday'] = df['Start_Time'].dt.day_name()
df['Is_Weekend'] = df['Weekday'].isin(['Saturday', 'Sunday']).astype(int)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(x='Hour', data=df, palette='viridis')
plt.title('Accidents by Hour of the Day', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=0)
plt.show()

##Insights:Peak hours for accidents (rush hours: morning & evening).Helps in road safety measures like better traffic control.
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Weekday', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='coolwarm')
plt.title('Accidents by Day of the Week', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=30)
plt.show()

## Insights: More accidents on weekdays or weekends? Helps traffic control planning on high-risk days.
```

```python
plt.figure(figsize=(14,6))
sns.countplot(y='Weather_Condition', data=df, order=df['Weather_Condition'].value_counts().index[:10], palette='coolwarm')
plt.title('Top 10 Weather Conditions Leading to Accidents', fontsize=14)
plt.xlabel('Number of Accidents', fontsize=12)
plt.ylabel('Weather Condition', fontsize=12)
plt.show()

## Insights: Which weather conditions cause most accidents? Can help authorities issue weather-specific road safety alerts.
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Severity', y='Temperature(F)', data=df, palette='magma')
plt.title('Impact of Temperature on Accident Severity', fontsize=14)
plt.xlabel('Accident Severity', fontsize=12)
plt.ylabel('Temperature (Â°F)', fontsize=12)
plt.show()

## Insights: Does extreme cold or heat increase accident severity?
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Severity', y='Visibility(mi)', data=df, palette='cividis')
plt.title('Impact of Visibility on Accident Severity', fontsize=14)
plt.xlabel('Accident Severity', fontsize=12)
plt.ylabel('Visibility (miles)', fontsize=12)
plt.show()

##Insights: Low visibility leads to higher accident severity. Helps in policy-making for foggy/rainy conditions.
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Severity', data=df, palette='inferno')
plt.title('Accident Severity Distribution', fontsize=14)
plt.xlabel('Severity Level', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.show()


##Insights: How many minor vs. major accidents occur?
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x=df['Hour'].value_counts().index, y=df['Hour'].value_counts().values, marker='o', color='red')
plt.title('Accidents by Hour of the Day (Peak Traffic Management)', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.grid()
plt.show()

## Takeaway: Peak traffic hours need better monitoring and regulations.
```

```python
df['Month'] = df['Start_Time'].dt.month
plt.figure(figsize=(12,6))
sns.lineplot(x=df['Month'].value_counts().index, y=df['Month'].value_counts().values, marker='o', color='green')
plt.title('Accidents by Month (Seasonal Trends)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Severity'], y=df['Humidity(%)'], palette='coolwarm')
plt.title('Impact of Humidity on Accident Severity', fontsize=14)
plt.xlabel('Severity Level', fontsize=12)
plt.ylabel('Humidity (%)', fontsize=12)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,6))
weather_time_pivot = df.pivot_table(index='Weather_Condition', columns='Hour', aggfunc='size', fill_value=0)
sns.heatmap(weather_time_pivot, cmap='Reds', annot=False)
plt.title('Accidents by Hour and Weather Condition', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Weather Condition', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,8))
df['Traffic_Signal'].value_counts().plot.pie(autopct='%1.1f%%', colors=['blue', 'red'], startangle=90, explode=[0, 0.1])
plt.title('Accidents Near Traffic Signals', fontsize=14)
plt.ylabel('')
plt.show()
```

```python
df['Weekend'] = df['Weekday'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')

plt.figure(figsize=(8,5))
sns.barplot(x=df['Weekend'].value_counts().index, y=df['Weekend'].value_counts().values, palette='plasma')
plt.title('Weekday vs. Weekend Accidents', fontsize=14)
plt.xlabel('Day Type', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.show()
```
